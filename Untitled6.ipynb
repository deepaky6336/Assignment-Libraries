```python
#Q1. Load the given retail sales dataset into a Pandas DataFrame and display the first 5 rows along with the dataset information to understand its structure.

# ANSWER:
import pandas as pd

# Load dataset
df = pd.read_csv('retail_dataset .csv')

# First 5 rows
print("First 5 Rows:")
print(df.head())

# Dataset Information
print("\nDataset Information:")
df.info()
```

```python
#Q2. Inspect the dataset and identify missing values in each column. Display the total number of missing values present in the dataset.

#ANSWER 2:
# Missing values in each column
print("Missing values in each column:")
print(df.isnull().sum())

# Total missing values in dataset
print("\nTotal Missing Values in Dataset:")
print(df.isnull().sum().sum())
```

```python
#Q3. Replace missing values in the Quantity and Price columns with their respective mean values using Pandas.

#ANSWER 2: 
# Replace missing values with mean

df['Quantity'] = df['Quantity'].fillna(df['Quantity'].mean())
df['Price'] = df['Price'].fillna(df['Price'].mean())

# Check again
print(df[['Quantity','Price']].isnull().sum())
```

```python
#Q4. Remove all rows where the Product Category or Region column contains missing values.

#ANSWER 4: 
# Remove rows with missing values in Product Category or Region

df = df.dropna(subset=['Product Category', 'Region'])

# Check missing values again
print(df[['Product Category','Region']].isnull().sum())
```

```python
print(df.shape)
```

```python
#Q5. Create a new column called Revenue using NumPy, where:
                                                          #Revenue = Quantity × Price

                                                                                     #Display the updated dataset with the new column.

#ANSWER 5: 
import numpy as np

# Create Revenue column
df['Revenue'] = np.multiply(df['Quantity'], df['Price'])

# Display updated dataset
print(df.head())
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

print(df)
```

```python
#Q6. Using NumPy, calculate the total revenue generated from all transactions in the dataset.

#ANSWER 6:
import numpy as np

# Calculate total revenue
total_revenue = np.sum(df['Revenue'])

print("Total Revenue Generated:", total_revenue)
```

```python
#Q7. Group the dataset by Product Category and calculate the total revenue generated by each category.

#ANSWER 7:
# Group by Product Category and calculate total revenue

category_revenue = df.groupby('Product Category')['Revenue'].sum().reset_index()

print(category_revenue)
```

```python
#Q8. Based on the revenue calculated for each product category, identify the top 3 categories with the highest revenue and the bottom 3 categories with the lowest revenue.

#ANSWER 8:
# Total revenue by category
category_revenue = df.groupby('Product Category')['Revenue'].sum()

# Top 3 highest revenue categories
top3 = category_revenue.sort_values(ascending=False).head(3)

# Bottom 3 lowest revenue categories
bottom3 = category_revenue.sort_values(ascending=True).head(3)

print("Top 3 Categories with Highest Revenue:")
print(top3)

print("\nBottom 3 Categories with Lowest Revenue:")
print(bottom3)
```

```python
#Q9. Group the dataset by Region and calculate the total revenue generated in each region. Identify the region with the highest revenue and the region with the lowest revenue.

#ANSWER 9:
# Group by Region and calculate total revenue

region_revenue = df.groupby('Region')['Revenue'].sum()

print("Revenue by Region:")
print(region_revenue)

# Region with Highest Revenue
highest_region = region_revenue.idxmax()
highest_value = region_revenue.max()

print("\nHighest Revenue Region:")
print(highest_region, "-", highest_value)

# Region with Lowest Revenue
lowest_region = region_revenue.idxmin()
lowest_value = region_revenue.min()

print("\nLowest Revenue Region:")
print(lowest_region, "-", lowest_value)
```

```python
#Q10. Extract the month from the Date column using Pandas and group the dataset by month to calculate the total monthly revenue. Also, calculate the mean, median, and standard deviation of the Revenue column using NumPy.

#ANSWER 10:
import pandas as pd
import numpy as np

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')

# Extract Month
df['Month'] = df['Date'].dt.month

# Monthly Revenue
monthly_revenue = df.groupby('Month')['Revenue'].sum()

print("Monthly Revenue:")
print(monthly_revenue)

# Revenue Statistics using NumPy
print("\nMean Revenue:", np.mean(df['Revenue']))
print("Median Revenue:", np.median(df['Revenue']))
print("Standard Deviation:", np.std(df['Revenue']))
```
